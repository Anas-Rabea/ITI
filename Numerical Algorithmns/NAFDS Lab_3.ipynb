```python
def fun1(x):
    return  x**3 - x**2 - 15*x + 1
```

```python
def secant(fun , x_0, xk_1  , epsilon ):
    x=[xk_1]
    x.append(x_0)
    for i in range(2,20):
        f_x = fun(x[i-1])
    #     print(f_x)
        fd_x = (fun(x[i-1]) - fun(x[i-2]))/(x[i-1] - x[i-2])
    #     print(fd_x)
        x.append(x[i-1] - f_x / fd_x)
    #     print(x)
        if np.linalg.norm(x[i] - x[i-1]) <= epsilon:
            print("Iteration: " ,i - 2)
            print(f"f_x() =  " , fun(x[i]))
            print("POINTS : ",x)
            print("root: ", x[i])
            break
```

```python
fd6 = lambda x: np.array([2*x[0] + x[1],x[0]+2*x[1]])
hessian2 = lambda x:np.matrix([[2,1],[1,2]])
def multi_newton_with_learningrate(x,fd,hessian,eps,lr):

    i=0
    while np.sqrt(fd(x)[0]**2+fd(x)[1]**2) >= eps:
        x = x - np.linalg.inv(hessian(x)) @ fd(x) * lr
#         print(x)
        x= x.getA1() # to flatten the matrix
#         print(x)
        i+=1
    return x ,i,fd(x)
```

```python
x_0 = np.array([10,10])
x ,i,gd= multi_newton_with_learningrate( x_0,fun2_d,hessian2,0.001,1)
print("iterations = ",i)
print("x = ",x)
print("gredient",gd)
```
